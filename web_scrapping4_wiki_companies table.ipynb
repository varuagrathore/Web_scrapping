```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
page=requests.get(url)
soup=BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
page.text[0:500]
```

```python
unique_tags = set()

for tag in soup.find_all():
    unique_tags.add(tag.name)

for tag_name in unique_tags:
    print(tag_name)
```

```python
tag_names = ['h1', 'tbody', 'cite', 'link', 'body', 'head', 'nav', 'main', 'table', 'small',
             'ul', 'span', 'b', 'a', 'br', 'sup', 'header', 'img', 'script', 'footer', 'ol',
             'button', 'th', 'h2', 'title', 'noscript', 'p', 'meta', 'li', 'label', 'style',
             'input', 'form', 'td', 'tr', 'html', 'i', 'div', 'caption']

for tag_name in tag_names:
    tags = soup.find_all(tag_name) # Limit to the first 20 occurrences
    for tag in tags:
        print(f"Tag: {tag.name}\nText Content: {tag.text.strip()}\n")
```

```python
for paragraph in soup.find_all('tbody'):
    print(paragraph.text)
```

```python
print(soup.find_all('tbody'))
```

```python
import csv
tbody_tag = soup.find_all('tbody')[1]

if tbody_tag:
    table_rows = tbody_tag.find_all('tr')
    data_list = []

    # Extract headers separately
    header_row = table_rows[0]
    headers = [header.text.strip() for header in header_row.find_all('th')]
    
    # Extract data rows
    for row in table_rows[1:]:
        columns = row.find_all('td')
        row_data = [column.text.strip() for column in columns]
        data_list.append(row_data)

    csv_filename = 'table_data.csv'

    with open(csv_filename, mode='w', newline='', encoding='utf-8') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(headers)
        writer.writerows(data_list)

    print(f"Table data has been saved to {csv_filename}")
else:
    print("No valid table found in the page.")
```

```python
table=soup.find_all('table')[1]
```

```python
world_titles=table.find_all('th')
```

```python
world_table_titles=[title.text.strip() for title in world_titles]
```
